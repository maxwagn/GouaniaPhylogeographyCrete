## 0. Setup

````python
from pathlib import Path
import itertools
import re
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl

try:
    import seaborn as sns
    HAS_SEABORN = True
except ImportError:
    HAS_SEABORN = False

from scipy import stats
from statsmodels.stats.multitest import multipletests

warnings.filterwarnings("ignore", category=RuntimeWarning)

PROJECT = Path.cwd().resolve()

DATA = PROJECT / "data"
SNAKEMAKE = DATA / "snakemake"
METADATA = DATA / "metadata"
EXTERNAL = DATA / "external"

FIGURES = PROJECT / "figures"
TABLES = PROJECT / "tables"

for p in [FIGURES, TABLES]:
    p.mkdir(parents=True, exist_ok=True)

SPECIES = ["orientalis", "hofrichteri"]

SPECIES_LABEL = {
    "orientalis": r"$G.$ orientalis",
    "hofrichteri": r"$G.$ hofrichteri",
}

POPS = ["Kythira", "NNE-Mades", "NW-Petres", "SE-Vatos", "SW-Plakias"]
CRETE_POPS = ["NNE-Mades", "NW-Petres", "SE-Vatos", "SW-Plakias"]
FOCAL_POP = "Kythira"

SNP_SUFFIX = "variants.tif1.sft1.pass.snps.biallelic"
ALLSITES_SUFFIX = "all_sites.tif1.sft1.no_indels.biallelic"
WINDOW_LABEL = "w60000.m100.s30000"

mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["figure.dpi"] = 120
````

## 2. Helper functions

````python
def clean_pop_name(x):
    if pd.isna(x):
        return x
    return str(x).replace("_", "-")

def infer_population_from_sample(sample):
    sample = str(sample)
    for pop in POPS:
        if pop in sample:
            return pop
    return "Unknown"

def savefig(name, tight=True):
    if tight:
        plt.tight_layout()
    png = FIGURES / f"{name}.png"
    svg = FIGURES / f"{name}.svg"
    plt.savefig(png, dpi=300, bbox_inches="tight")
    plt.savefig(svg, bbox_inches="tight")
    print(f"Saved: {png}")
    print(f"Saved: {svg}")

def load_het(species):
    path = SNAKEMAKE / "results" / "heterozygosity_genome" / f"{species}_{SNP_SUFFIX}.het"
    df = pd.read_csv(path, sep=r"\s+")
    if "INDV" in df.columns:
        df["sample"] = df["INDV"]
    df["species"] = species
    df["population"] = df["sample"].map(infer_population_from_sample)
    return df

def load_popgenwindows(species):
    path = SNAKEMAKE / "results" / "popgenwindows" / f"{species}_{ALLSITES_SUFFIX}.{WINDOW_LABEL}.Fst.Dxy.pi.dA.csv.gz"
    df = pd.read_csv(path, sep="\t", compression="gzip")
    df["species"] = species
    if {"scaffold", "start", "end"}.issubset(df.columns):
        df["window_id"] = df["scaffold"].astype(str) + ":" + df["start"].astype(str) + "-" + df["end"].astype(str)
    return df

def load_tajimas_d(species):
    path = SNAKEMAKE / "results" / "TajimasD" / f"{species}_{ALLSITES_SUFFIX}.{WINDOW_LABEL}.D.tsv.gz"
    df = pd.read_csv(path, sep="\t", compression="gzip")
    df["species"] = species
    if {"scaffold", "start", "end"}.issubset(df.columns):
        df["window_id"] = df["scaffold"].astype(str) + ":" + df["start"].astype(str) + "-" + df["end"].astype(str)
    return df

def load_af(species):
    path = SNAKEMAKE / "results" / "AF" / f"AF_{species}.tsv.gz"
    df = pd.read_csv(path, sep="\t", compression="gzip")
    df["species"] = species
    return df

def metric_columns(df, metric):
    return [c for c in df.columns if c.startswith(metric + "_")]

def pairwise_col(metric, pop1, pop2, columns):
    c1 = f"{metric}_{pop1}_{pop2}"
    c2 = f"{metric}_{pop2}_{pop1}"
    if c1 in columns:
        return c1
    if c2 in columns:
        return c2
    return None
````

## 4. Windowed population-genomic summaries

````python
pgw = {s: load_popgenwindows(s) for s in SPECIES}
for s, df in pgw.items():
    print(s, df.shape)
    display(df.head())
````

````python
def summarize_window_metrics(df, species):
    rows = []
    for metric in ["pi", "dxy", "dA", "Fst"]:
        cols = metric_columns(df, metric)
        for col in cols:
            rows.append({
                "species": species,
                "metric": metric,
                "comparison": col.replace(metric + "_", ""),
                "mean": df[col].mean(skipna=True),
                "median": df[col].median(skipna=True),
                "sd": df[col].std(skipna=True),
                "n_windows_non_na": df[col].notna().sum(),
            })
    return pd.DataFrame(rows)

metric_summary = pd.concat([summarize_window_metrics(df, s) for s, df in pgw.items()], ignore_index=True)
metric_summary.to_csv(TABLES / "popgenwindows_metric_summary.tsv", sep="\t", index=False)
metric_summary.head(20)
````

````python
def plot_pairwise_heatmap_from_summary(metric_summary, metric, species):
    sdf = metric_summary[(metric_summary["species"] == species) & (metric_summary["metric"] == metric)].copy()
    mat = pd.DataFrame(index=POPS, columns=POPS, dtype=float)

    for _, row in sdf.iterrows():
        comp = row["comparison"].split("_")
        if len(comp) >= 2:
            p1, p2 = comp[0], comp[1]
            if p1 in mat.index and p2 in mat.columns:
                mat.loc[p1, p2] = row["mean"]
                mat.loc[p2, p1] = row["mean"]

    fig, ax = plt.subplots(figsize=(5, 4))
    if HAS_SEABORN:
        sns.heatmap(mat.astype(float), annot=True, fmt=".4f", cmap="viridis", ax=ax)
    else:
        im = ax.imshow(mat.astype(float), aspect="auto")
        plt.colorbar(im, ax=ax)
        ax.set_xticks(range(len(mat.columns)))
        ax.set_xticklabels(mat.columns, rotation=45, ha="right")
        ax.set_yticks(range(len(mat.index)))
        ax.set_yticklabels(mat.index)

    ax.set_title(f"{SPECIES_LABEL.get(species, species)} mean {metric}")
    savefig(f"heatmap_{species}_{metric}")
    plt.show()

for sp in SPECIES:
    for metric in ["Fst", "dxy", "dA"]:
        plot_pairwise_heatmap_from_summary(metric_summary, metric, sp)
````

````python
def plot_pi_by_population(species):
    df = pgw[species]
    pi_cols = [c for c in df.columns if c.startswith("pi_")]
    long = df[pi_cols].melt(var_name="population", value_name="pi")
    long["population"] = long["population"].str.replace("pi_", "", regex=False)

    fig, ax = plt.subplots(figsize=(7, 4))
    if HAS_SEABORN:
        sns.boxplot(data=long, x="population", y="pi", ax=ax)
    else:
        long.boxplot(column="pi", by="population", ax=ax)
    ax.set_title(f"{SPECIES_LABEL.get(species, species)} windowed nucleotide diversity")
    ax.set_xlabel("")
    ax.set_ylabel("pi")
    plt.suptitle("")
    savefig(f"{species}_pi_by_population")
    plt.show()

for sp in SPECIES:
    plot_pi_by_population(sp)
````

````python
def summarize_dtrios(df, species):
    out = df.copy()
    out["species"] = species
    p_cols = [c for c in out.columns if c.lower() in ["p", "pvalue", "p_value", "p-value"] or "pvalue" in c.lower()]
    if p_cols:
        pcol = p_cols[0]
        out["p_BH"] = multipletests(out[pcol].astype(float), method="fdr_bh")[1]
    return out

if dtrios:
    dtrios_all = pd.concat([summarize_dtrios(df, sp) for sp, df in dtrios.items()], ignore_index=True)
    dtrios_all.to_csv(TABLES / "Dsuite_Dtrios_combined_with_BH.tsv", sep="\t", index=False)
    display(dtrios_all.head())
````

## 6. PCA and ADMIXTURE

````python
def load_pca(species, maf="0.15"):
    vec_path = SNAKEMAKE / "results" / "pca" / f"{species}_{SNP_SUFFIX}.pca.maf{maf}.eigenvec"
    val_path = SNAKEMAKE / "results" / "pca" / f"{species}_{SNP_SUFFIX}.pca.maf{maf}.eigenval"

    if not vec_path.exists():
        print("Missing:", vec_path)
        return None, None

    vec = pd.read_csv(vec_path, sep=r"\s+", header=None)
    if vec.shape[1] >= 4:
        cols = ["FID", "IID"] + [f"PC{i}" for i in range(1, vec.shape[1]-1)]
        vec.columns = cols[:vec.shape[1]]
        vec["sample"] = vec["IID"]
    else:
        vec.columns = ["sample"] + [f"PC{i}" for i in range(1, vec.shape[1])]
    vec["population"] = vec["sample"].map(infer_population_from_sample)
    vec["species"] = species

    vals = None
    if val_path.exists():
        vals = pd.read_csv(val_path, header=None, names=["eigenvalue"])

    return vec, vals

pca_data = {}
for sp in SPECIES:
    vec, vals = load_pca(sp, maf="0.15")
    if vec is not None:
        pca_data[sp] = (vec, vals)
        display(vec.head())
````

````python
def plot_pca(species, pcx="PC1", pcy="PC2"):
    vec, vals = pca_data[species]
    fig, ax = plt.subplots(figsize=(5, 4))
    if HAS_SEABORN:
        sns.scatterplot(data=vec, x=pcx, y=pcy, hue="population", ax=ax, s=60)
    else:
        for pop, sdf in vec.groupby("population"):
            ax.scatter(sdf[pcx], sdf[pcy], label=pop)
        ax.legend()
    ax.set_title(f"{SPECIES_LABEL.get(species, species)} PCA")
    ax.set_xlabel(pcx)
    ax.set_ylabel(pcy)
    savefig(f"{species}_PCA_{pcx}_{pcy}")
    plt.show()

for sp in pca_data:
    plot_pca(sp)
````

````python
def parse_admixture_cv(species):
    rows = []
    folder = SNAKEMAKE / "results" / "admixture"
    for path in sorted(folder.glob(f"admixture_{species}_{SNP_SUFFIX}.clusters*.out")):
        m = re.search(r"clusters(\d+)", path.name)
        k = int(m.group(1)) if m else None
        text = path.read_text(errors="ignore")
        cv = None
        m_cv = re.search(r"CV error.*?:\s*([0-9.eE+-]+)", text)
        if m_cv:
            cv = float(m_cv.group(1))
        rows.append({"species": species, "K": k, "CV_error": cv, "file": str(path)})
    return pd.DataFrame(rows)

cv = pd.concat([parse_admixture_cv(sp) for sp in SPECIES], ignore_index=True)
cv.to_csv(TABLES / "admixture_cv_errors.tsv", sep="\t", index=False)
cv
````

````python
if not cv.empty and cv["CV_error"].notna().any():
    fig, ax = plt.subplots(figsize=(5, 4))
    for sp, sdf in cv.groupby("species"):
        sdf = sdf.sort_values("K")
        ax.plot(sdf["K"], sdf["CV_error"], marker="o", label=SPECIES_LABEL.get(sp, sp))
    ax.set_xlabel("K")
    ax.set_ylabel("ADMIXTURE CV error")
    ax.legend()
    savefig("admixture_cv_error")
    plt.show()
else:
    print("No ADMIXTURE CV errors found yet.")
````

## 7. Genome scans: PBS, dA, Tajima’s D

````python
def fst_to_T(fst):
    fst = np.asarray(fst, dtype=float)
    fst = np.clip(fst, 0, 0.999999)
    return -np.log(1 - fst)

def compute_pbs(df, focal=FOCAL_POP, references=CRETE_POPS):
    out = df[["scaffold", "start", "end", "window_id"]].copy()
    for p1, p2 in itertools.combinations(references, 2):
        f_ab_col = pairwise_col("Fst", focal, p1, df.columns)
        f_ac_col = pairwise_col("Fst", focal, p2, df.columns)
        f_bc_col = pairwise_col("Fst", p1, p2, df.columns)
        if f_ab_col is None or f_ac_col is None or f_bc_col is None:
            print("Missing FST columns for trio:", focal, p1, p2)
            continue
        out[f"PBS_{focal}_{p1}_{p2}"] = (fst_to_T(df[f_ab_col]) + fst_to_T(df[f_ac_col]) - fst_to_T(df[f_bc_col])) / 2
    pbs_cols = [c for c in out.columns if c.startswith("PBS_")]
    out["PBS_min_across_trios"] = out[pbs_cols].min(axis=1, skipna=True)
    out["PBS_mean_across_trios"] = out[pbs_cols].mean(axis=1, skipna=True)
    return out

pbs = {sp: compute_pbs(pgw[sp]) for sp in SPECIES}
for sp, df in pbs.items():
    df.to_csv(TABLES / f"{sp}_PBS_windows.tsv", sep="\t", index=False)
    print(sp, df.shape)
````

````python
def classify_pbs_outliers(pbs_df, quantile=0.95):
    pbs_cols = [c for c in pbs_df.columns if c.startswith("PBS_")]
    out = pbs_df.copy()
    thresholds = {c: out[c].quantile(quantile) for c in pbs_cols}
    for c, t in thresholds.items():
        out[c + "_outlier"] = out[c] >= t
    outlier_cols = [c + "_outlier" for c in pbs_cols]
    out["PBS_outlier_all_trios"] = out[outlier_cols].all(axis=1)
    return out, thresholds

pbs_out = {}
pbs_thresholds = {}
for sp in SPECIES:
    pbs_out[sp], pbs_thresholds[sp] = classify_pbs_outliers(pbs[sp])
    pbs_out[sp].to_csv(TABLES / f"{sp}_PBS_outliers.tsv", sep="\t", index=False)
    print(sp, "PBS outliers:", int(pbs_out[sp]["PBS_outlier_all_trios"].sum()))
````

````python
def classify_da_outliers(df, focal=FOCAL_POP, crete_pops=CRETE_POPS, quantile=0.95):
    out = df[["scaffold", "start", "end", "window_id"]].copy()

    focal_cols = []
    for pop in crete_pops:
        col = pairwise_col("dA", focal, pop, df.columns)
        if col is not None:
            focal_cols.append(col)
            out[col + "_outlier"] = df[col] >= df[col].quantile(quantile)

    bg_cols = []
    for p1, p2 in itertools.combinations(crete_pops, 2):
        col = pairwise_col("dA", p1, p2, df.columns)
        if col is not None:
            bg_cols.append(col)
            out[col + "_background_outlier"] = df[col] >= df[col].quantile(quantile)

    focal_outlier_cols = [c + "_outlier" for c in focal_cols]
    bg_outlier_cols = [c + "_background_outlier" for c in bg_cols]

    out["dA_Kythira_vs_Crete_any_outlier"] = out[focal_outlier_cols].any(axis=1) if focal_outlier_cols else False
    out["dA_Crete_Crete_any_outlier"] = out[bg_outlier_cols].any(axis=1) if bg_outlier_cols else False
    out["dA_Kythira_specific_outlier"] = out["dA_Kythira_vs_Crete_any_outlier"] & ~out["dA_Crete_Crete_any_outlier"]
    return out

da_out = {}
for sp in SPECIES:
    da_out[sp] = classify_da_outliers(pgw[sp])
    da_out[sp].to_csv(TABLES / f"{sp}_dA_outliers.tsv", sep="\t", index=False)
    print(sp, "dA Kythira-specific outliers:", int(da_out[sp]["dA_Kythira_specific_outlier"].sum()))
````

````python
candidates = {}
for sp in SPECIES:
    merged = (
        pbs_out[sp][["scaffold", "start", "end", "window_id", "PBS_outlier_all_trios"]]
        .merge(da_out[sp][["window_id", "dA_Kythira_specific_outlier"]], on="window_id", how="inner")
    )
    merged["candidate_PBS_and_dA"] = merged["PBS_outlier_all_trios"] & merged["dA_Kythira_specific_outlier"]
    candidates[sp] = merged
    merged.to_csv(TABLES / f"{sp}_candidate_windows_PBS_dA.tsv", sep="\t", index=False)
    print(sp, "candidate windows:", int(merged["candidate_PBS_and_dA"].sum()))
````

````python
cand_sets = {sp: set(df.loc[df["candidate_PBS_and_dA"], "window_id"]) for sp, df in candidates.items()}
overlap = cand_sets["orientalis"] & cand_sets["hofrichteri"]

n_bg = len(set(candidates["orientalis"]["window_id"]) | set(candidates["hofrichteri"]["window_id"]))
n_or = len(cand_sets["orientalis"])
n_hof = len(cand_sets["hofrichteri"])
n_overlap = len(overlap)

p_hyper = stats.hypergeom.sf(n_overlap - 1, n_bg, n_or, n_hof) if n_bg else np.nan
expected = n_or * n_hof / n_bg if n_bg else np.nan
enrichment = n_overlap / expected if expected else np.nan

overlap_summary = pd.DataFrame([{
    "background_windows": n_bg,
    "orientalis_candidates": n_or,
    "hofrichteri_candidates": n_hof,
    "observed_overlap": n_overlap,
    "expected_overlap": expected,
    "enrichment": enrichment,
    "hypergeometric_p": p_hyper,
}])
overlap_summary.to_csv(TABLES / "convergent_candidate_window_overlap_summary.tsv", sep="\t", index=False)
overlap_summary
````

## 8. Tajima’s D and ΔTajima’s D

````python
tajd = {sp: load_tajimas_d(sp) for sp in SPECIES}
for sp, df in tajd.items():
    print(sp, df.shape)
    display(df.head())
````

````python
def find_tajd_columns(df):
    cols = {}
    for pop in POPS:
        matches = [c for c in df.columns if pop in c and re.search(r"(D|taj)", c, flags=re.IGNORECASE)]
        if matches:
            cols[pop] = sorted(matches, key=len)[0]
    return cols

def add_delta_tajd(df, cols, focal=FOCAL_POP, crete_pops=CRETE_POPS):
    out = df.copy()
    if focal not in cols:
        out["delta_TajD_Kythira_vs_Crete"] = np.nan
        return out
    crete_cols = [cols[p] for p in crete_pops if p in cols]
    if not crete_cols:
        out["delta_TajD_Kythira_vs_Crete"] = np.nan
        return out
    out["TajD_Kythira"] = out[cols[focal]]
    out["TajD_Crete_mean"] = out[crete_cols].mean(axis=1, skipna=True)
    out["delta_TajD_Kythira_vs_Crete"] = out["TajD_Kythira"] - out["TajD_Crete_mean"]
    return out

tajd_cols = {sp: find_tajd_columns(df) for sp, df in tajd.items()}
tajd_delta = {}
for sp in SPECIES:
    tajd_delta[sp] = add_delta_tajd(tajd[sp], tajd_cols[sp])
    tajd_delta[sp].to_csv(TABLES / f"{sp}_TajimasD_delta.tsv", sep="\t", index=False)
    print(sp, tajd_cols[sp])
````

````python
def test_candidate_tajd(species):
    td = tajd_delta[species]
    cand = candidates[species][["window_id", "candidate_PBS_and_dA"]]
    merged = td.merge(cand, on="window_id", how="left")
    merged["candidate_PBS_and_dA"] = merged["candidate_PBS_and_dA"].fillna(False)

    x = merged.loc[merged["candidate_PBS_and_dA"], "delta_TajD_Kythira_vs_Crete"].dropna()
    y = merged.loc[~merged["candidate_PBS_and_dA"], "delta_TajD_Kythira_vs_Crete"].dropna()

    if len(x) > 1 and len(y) > 1:
        stat, p = stats.mannwhitneyu(x, y, alternative="less")
    else:
        stat, p = np.nan, np.nan

    return {
        "species": species,
        "n_candidate": len(x),
        "n_background": len(y),
        "candidate_mean_delta_TajD": x.mean(),
        "background_mean_delta_TajD": y.mean(),
        "mannwhitney_U": stat,
        "p_candidate_more_negative": p,
    }

tajd_tests = pd.DataFrame([test_candidate_tajd(sp) for sp in SPECIES])
tajd_tests.to_csv(TABLES / "candidate_delta_TajimasD_tests.tsv", sep="\t", index=False)
tajd_tests
````

## 9. AF / SnpEff nearly fixed allele scans

````python
af = {sp: load_af(sp) for sp in SPECIES}
for sp, df in af.items():
    print(sp, df.shape)
    display(df.head())
````

````python
def af_col_for_population(df, pop):
    candidates = [c for c in df.columns if c == f"AF_{pop}" or c.endswith(f"_{pop}")]
    if candidates:
        return candidates[0]
    pop2 = pop.replace("-", "_")
    candidates = [c for c in df.columns if c == f"AF_{pop2}" or c.endswith(f"_{pop2}")]
    return candidates[0] if candidates else None

def classify_nearly_fixed_af(df, focal=FOCAL_POP, crete_pops=CRETE_POPS, high=0.95, low=0.05):
    out = df.copy()
    focal_col = af_col_for_population(out, focal)
    crete_cols = [af_col_for_population(out, p) for p in crete_pops]
    crete_cols = [c for c in crete_cols if c is not None]

    if focal_col is None or not crete_cols:
        print("Missing AF columns:", focal_col, crete_cols)
        out["nearly_fixed_reciprocal"] = False
        return out

    out["AF_Kythira"] = pd.to_numeric(out[focal_col], errors="coerce")
    out["AF_Crete_mean"] = out[crete_cols].apply(pd.to_numeric, errors="coerce").mean(axis=1, skipna=True)
    out["Kythira_high_Crete_low"] = (out["AF_Kythira"] >= high) & (out["AF_Crete_mean"] <= low)
    out["Kythira_low_Crete_high"] = (out["AF_Kythira"] <= low) & (out["AF_Crete_mean"] >= high)
    out["nearly_fixed_reciprocal"] = out["Kythira_high_Crete_low"] | out["Kythira_low_Crete_high"]
    return out

af_fixed = {}
for sp in SPECIES:
    af_fixed[sp] = classify_nearly_fixed_af(af[sp])
    af_fixed[sp].to_csv(TABLES / f"{sp}_nearly_fixed_AF_variants.tsv", sep="\t", index=False)
    print(sp, "nearly fixed reciprocal variants:", int(af_fixed[sp]["nearly_fixed_reciprocal"].sum()))
````

````python
def extract_gene_ids_from_ann(ann):
    if pd.isna(ann):
        return []
    genes = []
    for entry in str(ann).split(","):
        parts = entry.split("|")
        for idx in [3, 4]:
            if len(parts) > idx and parts[idx] and parts[idx] != ".":
                genes.append(parts[idx])
    return sorted(set(genes))

gene_sets_fixed = {}
for sp, df in af_fixed.items():
    if "ANN" not in df.columns:
        print("No ANN column in", sp)
        gene_sets_fixed[sp] = set()
        continue
    genes = set()
    for ann in df.loc[df["nearly_fixed_reciprocal"], "ANN"]:
        genes.update(extract_gene_ids_from_ann(ann))
    gene_sets_fixed[sp] = genes
    pd.Series(sorted(genes), name="gene").to_csv(TABLES / f"{sp}_nearly_fixed_AF_genes.txt", index=False, header=False)
    print(sp, len(genes), "genes")

fixed_gene_overlap = gene_sets_fixed.get("orientalis", set()) & gene_sets_fixed.get("hofrichteri", set())
pd.Series(sorted(fixed_gene_overlap), name="gene").to_csv(TABLES / "overlap_nearly_fixed_AF_genes.txt", index=False, header=False)
print("Shared nearly fixed AF genes:", len(fixed_gene_overlap))
````

## 10. Candidate genes near outlier windows

````python
GTF = EXTERNAL / "GCF_900634775.1_fGouWil2.1_genomic_chromnames.gtf"

def parse_gtf_attributes(attr):
    d = {}
    for part in str(attr).strip().split(";"):
        part = part.strip()
        if not part:
            continue
        if " " in part:
            key, val = part.split(" ", 1)
            d[key] = val.strip('"')
    return d

def load_gtf_genes(gtf_path):
    cols = ["scaffold", "source", "feature", "start", "end", "score", "strand", "frame", "attributes"]
    gtf = pd.read_csv(gtf_path, sep="\t", comment="#", names=cols, low_memory=False)
    genes = gtf[gtf["feature"].isin(["gene", "mRNA", "transcript"])].copy()
    attrs = genes["attributes"].map(parse_gtf_attributes)
    genes["gene_id"] = attrs.map(lambda d: d.get("gene_id") or d.get("ID") or d.get("transcript_id"))
    genes["gene_name"] = attrs.map(lambda d: d.get("gene_name") or d.get("Name") or d.get("gene"))
    return genes

if GTF.exists():
    genes_gtf = load_gtf_genes(GTF)
    print(genes_gtf.shape)
    display(genes_gtf.head())
else:
    genes_gtf = None
    print("GTF not found:", GTF)
````

````python
def genes_near_windows(windows, genes, flank=30000):
    if genes is None:
        return pd.DataFrame()
    rows = []
    for _, w in windows.iterrows():
        scaffold = str(w["scaffold"])
        mid = int((w["start"] + w["end"]) / 2)
        lo = mid - flank
        hi = mid + flank
        g = genes[
            (genes["scaffold"].astype(str) == scaffold) &
            (genes["end"] >= lo) &
            (genes["start"] <= hi)
        ].copy()
        for _, gene in g.iterrows():
            rows.append({
                "window_id": w["window_id"],
                "scaffold": scaffold,
                "window_start": w["start"],
                "window_end": w["end"],
                "query_start": lo,
                "query_end": hi,
                "gene_id": gene.get("gene_id"),
                "gene_name": gene.get("gene_name"),
                "gene_start": gene["start"],
                "gene_end": gene["end"],
            })
    return pd.DataFrame(rows)

for sp in SPECIES:
    cand = candidates[sp].loc[candidates[sp]["candidate_PBS_and_dA"]].copy()
    genes = genes_near_windows(cand, genes_gtf, flank=30000)
    genes.to_csv(TABLES / f"{sp}_candidate_window_genes_pm30kb.tsv", sep="\t", index=False)
    print(sp, genes.shape)
````

### 11.1 dXY-only outlier checks

````python
def classify_dxy_outliers(df, focal=FOCAL_POP, crete_pops=CRETE_POPS, quantile=0.95):
    out = df[["scaffold", "start", "end", "window_id"]].copy()
    cols = []
    for pop in crete_pops:
        col = pairwise_col("dxy", focal, pop, df.columns)
        if col:
            cols.append(col)
            out[col + "_outlier"] = df[col] >= df[col].quantile(quantile)
    out["dxy_Kythira_vs_Crete_any_outlier"] = out[[c + "_outlier" for c in cols]].any(axis=1) if cols else False
    return out

dxy_out = {sp: classify_dxy_outliers(pgw[sp]) for sp in SPECIES}
for sp, df in dxy_out.items():
    df.to_csv(TABLES / f"APPENDIX_{sp}_dxy_outliers.tsv", sep="\t", index=False)
    print(sp, int(df["dxy_Kythira_vs_Crete_any_outlier"].sum()))
````

### 11.2 pi versus outlier statistics

````python
def plot_pi_vs_stat(species, stat="dA", pop_a=FOCAL_POP, pop_b="NNE-Mades"):
    df = pgw[species]
    pi_col = f"pi_{pop_a}"
    stat_col = pairwise_col(stat, pop_a, pop_b, df.columns)
    if pi_col not in df.columns or stat_col is None:
        print("Missing columns", pi_col, stat_col)
        return
    fig, ax = plt.subplots(figsize=(4.5, 4))
    ax.scatter(df[pi_col], df[stat_col], s=8, alpha=0.4)
    ax.set_xlabel(pi_col)
    ax.set_ylabel(stat_col)
    ax.set_title(f"{SPECIES_LABEL.get(species, species)}: {stat_col} vs {pi_col}")
    savefig(f"APPENDIX_{species}_{stat_col}_vs_{pi_col}")
    plt.show()

for sp in SPECIES:
    plot_pi_vs_stat(sp, stat="dA", pop_b="NNE-Mades")
````
